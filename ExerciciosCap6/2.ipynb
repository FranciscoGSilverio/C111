```python
# 2. Por meio do dataset space.csv, trace um gráfico em barras
# mostrando quantas empresas espaciais diferentes os EUA e a CHINA
# possuem;
# Dica: não se esqueça de retirar os resultados repetidos
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('space.csv', encoding='utf-8', delimiter=';')
df_unique = df.drop_duplicates(subset=['Company Name'])

df_usa = df_unique[df_unique['Location'].str.contains('USA')]
df_china = df_unique[df_unique['Location'].str.contains('China')]

df_count_usa = df_usa.count()
df_count_china = df_china.count()

plt.bar(['USA', 'China'], [df_count_usa['Company Name'], df_count_china['Company Name']])
plt.xlabel('País')
plt.ylabel('Número de Empresas Espaciais')
plt.title('Número de Empresas Espaciais por País')
plt.show()




```
