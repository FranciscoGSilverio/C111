```python
# Por meio do dataset paises.csv, trace dois gráficos de linhas em
# um mesmo plano cartesiano, um mostrando a taxa de mortalidade (Deathrate) e outro a taxa de natalidade (Birthrate) dos países da
# América do Norte;

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('paises.csv', on_bad_lines='skip', encoding='latin1', delimiter=';', encoding_errors='ignore')

norte = df[df['Region'].str.contains('NORTHERN AMERICA')]
countries_norte = norte['ï»¿Country'].tolist()

plt.plot(norte['ï»¿Country'], norte['Deathrate'], label='Taxa de Mortalidade', color='blue')
plt.plot(norte['ï»¿Country'], norte['Birthrate'], label='Taxa de Natalidade', color='red')
plt.xlabel('Países')
plt.ylabel('Taxa')
plt.title('Taxa de Mortalidade e Natalidade dos Países da América do Norte')
plt.legend()


```
